Целевая метрика уже выбрана за нас: мы будем считать $\text{ROC-AUC}$, но не простой, а коэффициент Джини:

$$ \text{Gini} = 2 \cdot \text{ROC-AUC} - 1$$

Конечная цель данного мероприятия — собрать пайплайн машинного обучения от и до, начиная с предобработки данных, заканчивая оптимизацией. При желании это можно доделать до целого пет-проекта, особенно если добавить сбор данных и деплой модели, но в дз этого не будет :(.

```python
from sklearn.metrics import roc_auc_score


def gini(y_true, y_score):
    return 2 * roc_auc_score(y_true, y_score) - 1.0
```

```python
from scipy.sparse import csr_array

class HeroesEncoder:

    def fit(self, X, y=None):
        self.id_heroes = X['hero_id'].unique()
        self.id_heroes_idx = {j: i for i, j in enumerate(self.id_heroes)}
        return self

    def transform(self, X, y=None):
        # ヾ(⌐■_■)ノ♪ your code here
        r = []
        c = []
        d = []

        for m, g in X.groupby('match_id'):
            for _, ro in g.iterrows():
                r.append(m)
                c.append(self.id_heroes_idx[ro['hero_id']])
                if 0 <= ro['player_slot'] <= 4:
                    d.append(1)
                else:
                    d.append(-1)

        id_matches = X['match_id'].unique()
        id_matches_idx = {j: i for i, j in enumerate(id_matches)}
        r = [id_matches_idx[i] for i in r]

        m = csr_array((d, (r, c)), shape=(len(id_matches), len(self.id_heroes)))
        return m, id_matches
```
